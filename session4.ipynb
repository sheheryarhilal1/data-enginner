```python
firstName = ["Ali","Abid","Adnan"]
lastName =  ["Ahmed","Anjum","Azad"]
for first in firstName:
    for last in lastName:
        print(first + " " + last)
```

```python
cities_to_visit = []
user_input = ''
while user_input != 'quit':
    city = input('Enter city to visit:')
    if city =='quit':
        user_input = 'quit'
    else:
        cities_to_visit.append(city)
        
print(cities_to_visit)
```

```python
# pop should remove the first value inside a set arrange value as well and remove value sequence wise 
s = {6, 7, 3, 4}

removed = s.pop()
print("Removed:", removed)
print("Set:", s)
```

```python
# union should merge two variable in a new variable we write c= a.union(b)
a = {1, 2, 3}
b = {3, 4, 5}

c = a.union(b)
print(c)
```

```python
# symmetric_difference should also use between two variable it remove same values on both variable and provide remaining value in new variable we write a.symmetric_difference(b)
a = {1, 2, 3}
b = {3, 4, 5}

c = a.symmetric_difference(b)
print(c)
```
