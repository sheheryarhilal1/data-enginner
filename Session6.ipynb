```python
def calculate_salary(basic_salary,*allowance,**deductions):
    total_allowance = sum(allowance)
    total_deduction = sum(deductions.values())

    gross_salary =basic_salary+total_allowance
    net_salary = gross_salary-total_deduction

    print("Basic Salary:",basic_salary)
    print("Total Allowance:",total_allowance)
    print("Total Deduction:",total_deduction)
    print("Net Salary:",net_salary)

    return net_salary
```

```python
calculate_salary(40000,2000,1000,x=3000,y=4000)
```

 **Order Calculator**

```python
def calculator_order(price,quantity,tax=0.05,discount=0,shipping=100):
    """price:product price
    quantity:number of item
    tax:tax percentage(default 5%)
    discount:discount amount (default 0)
    shipping:shipping charges (default 100)"""

    subtotal = price*quantity
    tax_amount=subtotal*tax
    total= subtotal+tax_amount+shipping-discount
    print(subtotal)
    print(tax_amount)

    return total
```

```python
calculator_order(5000,quantity=5)
```

**Employee salary calculator** 

```python
def calculate_salarys(basic_salary,bonus=0,overtime_hour=0,tax_rates=0.1):
    """
    basic_salary:employee base salary
    bonus:extra bonus amount
    overtime_hour:extra working hours
    tax_rates:tax percentage(default 10%)
    """

    over_time_pay=overtime_hour*500 #500 per hour
    gross_salary=basic_salary+bonus+over_time_pay
    tax_amount=gross_salary*tax_rates
    net_salary=gross_salary-tax_amount

    return net_salary
    
```

```python
calculate_salarys(100000,overtime_hour=3)
```

```python
def check_result(marks):

    def is_pass():
        return marks>=40

    if is_pass():
            return "Pass"
    else:
            return "Fail"
```
