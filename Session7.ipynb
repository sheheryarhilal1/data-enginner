**INHERITANCE**

Inheritance is  a fundamental concept in object oriented programming where a new class called the child
class or sub class is derived from an existing class known as the parent classes or superclasses The child class inherit
the attributes and method of the parent  class allowing for code reusability and the ability to extend or  modify  the 
behaviour of  the parent class without  altering its original code

**key Point**

*purpose :* to promote reusability,maintainability,and scalability of code
*Access :* The child class can be use or over ride the method and properties of the parent class
parent class-> base class,super class
child class-> derived class,subclass

```python
class Employee:
    def __init__(self,name,Employee_id,salary):
        self.name = name
        self.Employee_id =  Employee_id
        self.salary =  salary
    def work(self):
        print(f'{self.name} is working')
    def get_details(self):
            print(f'name:{self.name}.Employee_id:{self.Employee_id},salary{ self.salary}')
        
```

```python
employee = Employee('waqar',101,50000)
```

```python
employee.name
```

```python
employee.get_details()
```

```python
employee.Employee_id
```

```python
employee.salary
```

```python
employee.work()
```
