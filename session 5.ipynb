```python
# for showing keys we use .keys(): 
abc ={"S No." : 1,
     "question": "what is your country name",
     "option" :["Pakistan","India","China","Iran"],
     "answer": "Pakistan"
    }

for each_key in abc.keys():
    print(each_key)
```

```python
# for showing values we use .values(): 
abc = {
    "S No." : 1,
    "question": "what is your country name",
    "option" :["Pakistan","India","China","Iran"],
    "answer": "Pakistan"
}

for value in abc.values():
    print(value)
```

```python
quiz = [
    {"S No." : 1,
     "question": "what is your country name",
     "option" :["Pakistan","India","China","Iran"],
     "answer": "Pakistan"
    },
    {
     "S No." : 2,
     "question": "what is your city name",
     "option" :["Karachi","Islamabad","Quetta","Sukkur"],
     "answer": "Karachi"
    },
    {
     "S No." : 3,
     "question": "what is your name",
     "option" :["ali","ahmed","waqar","saad"],
     "answer": "saad"
    },
    {
     "S No." : 4,
     "question": "what is your id",
     "option" :["123","456","789","345"],
     "answer": "456"
    }
]

score = 0

for quizApp in quiz:
    data = str(quizApp["S No."]) + " " + quizApp["question"] + "\n"
    
    for opt in quizApp["option"]:
        data += opt + "\n"
    
    ans = input(data)
    
    if ans == quizApp["answer"]:
        score += 1

print("Your Score:", score)
```
